```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

dados_amz = pd.read_csv('dados/linguagens_amzn.csv')
dados_apple = pd.read_csv('dados/linguagens_apple.csv')
dados_netflix = pd.read_csv('dados/linguagens_netflix.csv')
dados_spotify = pd.read_csv('dados/linguagens_spotify.csv')

```

```python
dados_amz = dados_amz.fillna('not_identified')
dados_netflix = dados_netflix.fillna('not_identified')
dados_spotify = dados_spotify.fillna('not_identified')
dados_apple = dados_apple.fillna('not_identified')
```

```python
dados_amz['linguagem'].count()
contagem_amz = dados_amz['linguagem'].value_counts()
```

```python
plt.figure(figsize=(10, 6)) 
contagem_amz.plot(kind='bar', color='skyblue')  

# Personalize o gráfico
plt.title('Contagem de Linguagens no Repositório Amazon')  
plt.xlabel('Linguagem')  
plt.ylabel('Contagem')  

# Exibe o gráfico
plt.xticks(rotation=45) 
plt.tight_layout() 
plt.show()  






```

```python
dados_netflix.count()
contagem_netflix = dados_netflix['linguagem'].value_counts()

plt.figure(figsize=(10, 6)) 
contagem_netflix.plot(kind='bar', color='green')  

# Personalize o gráfico
plt.title('Contagem de Linguagens no Repositório Netflix')  
plt.xlabel('Linguagem')  
plt.ylabel('Contagem')  

# Exibe o gráfico
plt.xticks(rotation=45) 
plt.tight_layout() 
plt.show()  
```

```python
dados_apple.count()
contagem_apple = dados_apple['linguagem'].value_counts()

plt.figure(figsize=(10, 6)) 
contagem_apple.plot(kind='bar', color='skyblue')  

# Personalize o gráfico
plt.title('Contagem de Linguagens no Repositório Apple')  
plt.xlabel('Linguagem')  
plt.ylabel('Contagem')  

# Exibe o gráfico
plt.xticks(rotation=45) 
plt.tight_layout() 
plt.show()  
```

```python
dados_spotify.count()
contagem_spotify = dados_spotify.value_counts()

plt.figure(figsize=(10, 6)) 
contagem_netflix.plot(kind='bar', color='skyblue')  

# Personalize o gráfico
plt.title('Contagem de Linguagens no Repositório Spotify')  
plt.xlabel('Linguagem')  
plt.ylabel('Contagem')  

# Exibe o gráfico
plt.xticks(rotation=45) 
plt.tight_layout() 
plt.show()  
```
